```python
# install chromium, its driver, and selenium
!apt update
!apt install chromium-chromedriver
!pip install selenium
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# open it, go to a website, and get results
wd = webdriver.Chrome(options=options)
```

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import requests
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from PIL import Image
wd.get("https://monkeytype.com/")
try:
  element = wd.find_element(By.CLASS_NAME, "acceptAll").click()
except:
  start = time.time()
  end = time.time()
  while end-start < 5:
    pic = 0
    res = 0
    elem1 = wd.find_element(By.ID, "words")
    # time.sleep(5)
    # print(elem1.get_attribute("innerHTML"))
    inputt = wd.find_element(By.ID, "wordsInput")
    first_word = ""
    inner = elem1.get_attribute("innerHTML")
    words = elem1.find_elements(By.XPATH,"./child::*")
    count = []
    for j in words:
      try:
        word_innerhtml = j.get_attribute("innerHTML")
        print(word_innerhtml)
      except:
        wd.save_screenshot("idk1.png")
        break
      active = ""
      s1 = "<letter>"
      for k in range(len(word_innerhtml)):
        if word_innerhtml[k:k+len(s1)] == s1:
          active += word_innerhtml[k+len(s1)]
      try:
        inputt.send_keys(active + " ")
      except:
        wd.save_screenshot("idk1.png")
        break

      count.append(active)
      piclink = "mon" + str(pic) + ".png"
      wd.save_screenshot(piclink)
      pic += 1
      end = time.time()
  time.sleep(10) 
  reslink = "result" + str(res) + ".png"
  wd.save_screenshot(reslink)
  res += 1
  end = time.time()
  print("End - start: ", end-start)
  print("Len of count: ", len(count))
  print(count)

    


  # s1 = "<letter>"
  # for j in range(len(inner)):
  #   if inner[j:j+len(s1)] == s1:
  #     first_word += inner[j+len(s1)]
  #     if len(first_word) == 3:
  #       break 
  # time.sleep(1)
  # inputt.send_keys(first_word)
```
